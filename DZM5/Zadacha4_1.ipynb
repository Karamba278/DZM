```python
%matplotlib inline

import numpy as np

print('В игре в рулетку будет 10 попыток, что бы было интереснее!')
x1=int(input('Ваша ставка? (от 0 до 36)'))
for i in range(0, 10):
    x = np.random.uniform(0, 36)
    x=round(x)
    if x==x1:
        print('Ура! Вы победили! Выпало поле ',x )
        break
    else:
        print(f'В этот раз Вам не повезло. Выпало поле {x}. Вам обязательно повезет в следующий раз!')
```
