```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) +x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) +x - x*y - 1)

x1, y1 =  fsolve(equations, (-2, 1))
              
print (x1, y1)
```
